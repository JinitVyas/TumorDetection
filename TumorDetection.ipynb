# **Training The Data Set**

```python
# divide the dataset into train and test set
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test  = train_test_split(x,y, test_size=0.5) # here x and y are the data and the prediction value
```

```python
df.shape
```



---
The shape of features variables for train and test


```python
x_train.shape # the training_input part (70%) of the df
```

```python
x_test.shape # the test_input part (30%) of the df
```



---
The shape of **diagnosis column** variables for train and test


```python
y_train.shape # the training output part (70%) of df['diagnosis']
```

```python
y_test.shape # the test output part (30%) of df['diagnosis'] FOR VERIFICATION after the prediction
# also known as actualValue/GroundTruth/TargetValues
```

```python
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)

x_train
```

```python
# Apply random forst classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier # A technique that predicts on the voting from different decision trees nd highest voted outcome becomes the final prediction

rfc = RandomForestClassifier()

# Defining how the data curve looks like
rfc.fit(x_train,y_train)

# Predicting from the x_test(the test part_30% data_input columns)
y_pred = rfc.predict(x_test)

# and finally printing the accuracy by passing the y_test(GroundTruth) and the y_pred(values predicted from RandomForestClassifier)
print(accuracy_score(y_test, y_pred))
```
